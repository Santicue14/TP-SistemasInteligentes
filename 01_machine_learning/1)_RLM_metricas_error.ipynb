# Regresión Lineal Múltiple y Métricas de Error
**Dataset:** Costo Seguro Médico provisto en campus virtual. Predice el costo del seguro médico de una
persona en función de su edad, género, IMC (BMI), cantidad de hijos y si fuma o no.
**Dataset original:** https://www.kaggle.com/datasets/mirichoi0218/insurance
**Tarea:** El objetivo es predecir el costo médico en función de sus características.


A continuación se importan las librerías a utilizar sobre este punto

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

Se importa la fuente datos (dataset) tener en cuenta que esta es nuestra verdad fundamental. Se listan los primeros 5 datos para verificar

```python
dataset = pd.read_csv("dataset_costo_seguro_medico.csv")

dataset.head()
```

Debajo, definimos nuestras X e y y luego lo separamos con la variable prc_dataset_test a cuánto de ese dataset tomaremos como prueba y cuanto como datos de entrenamiento

```python
X = dataset.iloc[:,:-1] # Todas las filas, todas las columnas menos la última
y = dataset.iloc[:,-1] # Todas las filas, solo última columna

prc_dataset_test = 0.2 # 20 %

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=prc_dataset_test)
```

Debajo, instanciamos el modelo y lo entrenamos con justamente los datos de entrenamiento. Usaremos regresión lineal ya que, la consigna lo indica y nuestro objetivo es predecir un valor continuo.

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

Debajo predecimos con los datos de prueba, y evaluamos el MAE y el MSE de cada uno.

```python
MAE = mean_absolute_error(y_test, model.predict(X_test))
MSE = mean_squared_error(y_test, model.predict(X_test))

print(f"\nMAE: {MAE}")
print(f"MSE: {MSE}")
```

Aplicamos StandardScaler para estandarizar las características de nuestro dataset.

```python
scaler = StandardScaler()

# Ajustar el escalador con los datos de entrenamiento y transformar X_train
X_train_scaled = scaler.fit_transform(X_train)

# Transformar X_test usando el mismo escalador ajustado
X_test_scaled = scaler.transform(X_test)
```

Con los datos escalados, entrenamos otro modelo con `X_train_scaled` y evaluar con `X_test_scaled` para ver si mejora el rendimiento.

```python
# Reentrenar el modelo con los datos escalados
model_scaled = LinearRegression()
model_scaled.fit(X_train_scaled, y_train)
```

Predecimos las Y pero con los datos y modelo escalados, luego comparamos MAE y MSE del original vs. los datos escalados

```python
# Realizar predicciones con los datos de prueba escalados
y_pred_scaled = model_scaled.predict(X_test_scaled)

# Evaluar el rendimiento del modelo escalado
MAE_scaled = mean_absolute_error(y_test, y_pred_scaled)
MSE_scaled = mean_squared_error(y_test, y_pred_scaled)

print(f"MAE (escalado) {MAE_scaled} vs MAE (original): {MAE}: {MAE_scaled-MAE}")

print(f"MSE (escalado) {MSE_scaled} vs MSE (original): {MSE}: {MSE_scaled-MSE}")
```

## Respuestas:

### ¿Qué significa el valor del MAE?
El valor del MAE es el valor de error absoluto, este nos indica un promedio de por cuánto se equivoca nuestro modelo y el valor nos lo indica en la misma unidad de medida de nuestra y, que en este caso es charges (o sea en moneda) Podemos decir, que se está equivocando por $4.100 a lo que en realidad se da hoy. Este valor es la distancia al valor original, ya sea negativo o positivo.

### ¿Por qué el MSE es numéricamente mucho más grande que el MAE en este caso?
Porque el MSE castiga al error elevándolo al ^2. Osea si mi modelo falla prediciendo los datos de prueba por una distancia de 4000, se eleva al cuadrado dando un error de 16 millones.


### ¿Mejora algo al escalar características? ¿Por qué?
No, no mejora notablemente, como vemos tenemos una reducción mínima del MAE, esto se debe a que el algoritmo de la regresión lineal, ya compensa matemáticamente el cambio de escala ajustando los coeficientes. De igual manera es fundamental ya que nos permite comparar los pesos reales de cada variable frente a la consigna.
